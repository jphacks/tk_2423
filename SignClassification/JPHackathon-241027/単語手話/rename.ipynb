```python
import os

def rename_jpg_files_in_all_subfolders(root_folder_path):
    # 指定したフォルダ内のすべてのサブフォルダを探索
    for folder_name in os.listdir(root_folder_path):
        # サブフォルダのパスを取得
        subfolder_path = os.path.join(root_folder_path, folder_name)

        # サブフォルダかどうか確認
        if os.path.isdir(subfolder_path):
            # サブフォルダ内のJPGファイルを取得
            jpg_files = [f for f in os.listdir(subfolder_path) if f.lower().endswith('.jpg')]

            # 各JPGファイルをリネーム（インデックスを3桁で表示）
            for index, file_name in enumerate(jpg_files, start=1):
                # 新しいファイル名を生成（3桁のゼロパディング）
                new_name = f"{folder_name}-{str(index).zfill(3)}-ikami.jpg"

                # 元ファイルのパスと新ファイルのパスを設定
                old_file_path = os.path.join(subfolder_path, file_name)
                new_file_path = os.path.join(subfolder_path, new_name)

                # リネーム実行
                os.rename(old_file_path, new_file_path)
                print(f"Renamed '{file_name}' in '{folder_name}' to '{new_name}'")
```

```python
# 使用例
rename_jpg_files_in_all_subfolders('/content/drive/Shareddrives/JPHackathon-241027/単語手話/伊神')
```

```python
import os
import shutil

def copy_jpg_files_without_renaming(root_folder_path, output_folder_path):
    # 出力フォルダが存在しなければ作成
    if not os.path.exists(output_folder_path):
        os.makedirs(output_folder_path)

    # 指定したフォルダ内のすべてのサブフォルダを探索
    for folder_name in os.listdir(root_folder_path):
        subfolder_path = os.path.join(root_folder_path, folder_name)

        # サブフォルダかどうか確認
        if os.path.isdir(subfolder_path):
            # サブフォルダ内のJPGファイルを取得
            jpg_files = [f for f in os.listdir(subfolder_path) if f.lower().endswith('.jpg')]

            # 各JPGファイルを出力フォルダに移動
            for file_name in jpg_files:
                old_file_path = os.path.join(subfolder_path, file_name)
                new_file_path = os.path.join(output_folder_path, file_name)

                # ファイル名が重複している場合は末尾に番号を追加
                counter = 1
                while os.path.exists(new_file_path):
                    base, ext = os.path.splitext(file_name)
                    new_file_path = os.path.join(output_folder_path, f"{base}_{counter}{ext}")
                    counter += 1

                # ファイルを移動
                shutil.copy2(old_file_path, new_file_path)
                print(f"Copy '{file_name}' from '{folder_name}' to '{new_file_path}'")
```

```python
# 使用例
copy_jpg_files_without_renaming('/content/drive/Shareddrives/JPHackathon-241027/単語手話/伊神', '/content/drive/Shareddrives/JPHackathon-241027/単語手話/伊神-まとめ')
```
